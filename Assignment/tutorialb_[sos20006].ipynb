```python
df = sns.load_dataset('iris')
setosa = df.loc[df.species=='setosa', "sepal_length"]
virginica = df.loc[df.species=='virginica', "sepal_length"]

plt.subplots(figsize=(7,6), dpi=100)
# [add code here]
sns.distplot(setosa,label="setosa")
sns.distplot(virginica,label="virginica")
#sns.regplot(x='septal_length')
# [add code here]
plt.title("Iris Histogram")
plt.legend()
plt.show()

plt.show()
```
